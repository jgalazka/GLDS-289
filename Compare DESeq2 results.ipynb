```python
import pandas as pd
import numpy as np
```

# Compare AdjP < 0.01, no LFC cutoff

```python
rna = pd.read_csv('Space_uG_over_Space_1G_rRNA.csv', index_col=0)
rna['logpvalue'] = -np.log10(rna['pvalue'])
rna.head()
```

```python
norna = pd.read_csv('Space_uG_over_Space_1G_no_rRNA.csv', index_col=0)
norna['logpvalue'] = -np.log10(norna['pvalue'])
norna.head()
```

```python
joined = rna.join(norna, how='inner', lsuffix='_RNA', rsuffix='_noRNA')
joined.head()
```

```python
rnaDE = rna[rna['padj'] < 0.1]
rnaDE.shape
```

```python
nornaDE = norna[norna['padj'] < 0.1]
nornaDE.shape
```

```python
joined['DE_RNA'] = joined['padj_RNA'] < 0.1
joined['DE_noRNA'] = joined['padj_noRNA'] < 0.1
joined.head()
```

```python
joined['Compare'] = joined['DE_RNA'] == joined['DE_noRNA']
joined.head()
```

```python
joined.to_csv('DESeq2_results_combined_p1_no_lfc.csv')
```
